## 1. Introduction: The Testing Burden Problem

A comprehensive LD assessment often requires dozens of observations across distinct domains. For a teacher, filling out long, fixed-length forms for multiple students is administratively burdensome. The **Taranga Adaptive Engine** solves this by dynamically routing questions based on early indicators.

### 1.1 Methodology: HRAT Algorithm
Taranga uses a custom algorithm that reduces question volume while preserving diagnostic coverage:

1. **Phase 1 (Gateway):** All students start with 5 "gateway" questions (1 per LD). 
2. **Suspicion Sorting:** Based on scores (1–5), domains are ranked by "suspicion level."
3. **Phase 2 (Deep-Dive):** The remaining 15 questions are allocated disproportionately to the highest-suspicion domains.
4. **Smart Imputation:** For low-suspicion domains where deep-dives were skipped, the engine imputes values based on the gateway baseline.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_theme(style="whitegrid", palette="muted")

# Constants for Analysis
FIXED_LENGTH = 60
ADAPTIVE_LENGTH = 20
```

## 2. Efficiency Benchmark

The chart below compares the time/effort burden of a traditional fixed screening vs. Taranga's adaptive approach. By focusing only on relevant domains, we achieve a **66% reduction** in question volume.

```python
labels = ['Traditional (Fixed)', 'Taranga (Adaptive)']
values = [FIXED_LENGTH, ADAPTIVE_LENGTH]

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=labels, y=values, palette="viridis")
plt.ylabel("No. of Observation Items per Student")
plt.title("Screening Efficiency Gain: Adaptive vs. Fixed-Length Protocols", fontsize=15, fontweight='bold')

# Add data labels
for i, v in enumerate(values):
    ax.text(i, v + 1, str(v), color='black', ha='center', fontweight='bold')

plt.show()
```

## 3. Allocation Strategy Logic

The adaptive engine allocates its 15 deep-dive questions based on current suspicions. Below is the simulation showing the question budget distribution for a student suspected of **Dyslexia and APD**.

```python
categories = ['Gateway (All)', 'High Suspicion (Rank 1)', 'High Suspicion (Rank 2)', 'Moderate (Rank 3)', 'Confirmatory (Rank 4/5)', 'Fill Items']
questions = [5, 4, 4, 2, 2, 3]

plt.figure(figsize=(12, 6))
sns.lineplot(x=categories, y=np.cumsum(questions), marker='o', sort=False, color='#6366F1')
plt.fill_between(categories, np.cumsum(questions), alpha=0.2, color='#6366F1')
plt.title("Cumulative Question Load Across Screening Phases", fontsize=14, fontweight='bold')
plt.ylabel("Cumulative Questions Asked")
plt.show()

print("PHASE ALLOCATION BREAKDOWN")
print("-" * 30)
for cat, q in zip(categories, questions):
    print(f"{cat:<25} | {q} items")
```
