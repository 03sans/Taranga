# AI Model Evaluation Report
This notebook provides the evaluation metrics for the Taranga AI model, including Classification Reports and Confusion Matrices for the 5 primary LD domains.

```python
import joblib
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, multilabel_confusion_matrix, hamming_loss
import os

# Set relative paths from backend/notebooks/
MODEL_PATH = '../app/ai/models/ml_model.joblib'
DATA_PATH = '../data/synthetic_dataset.csv'

FEATURE_COLS = [
    "q1_reading_pace", "q2_spelling_errors", "q3_letter_reversal", "q4_phonological",
    "q5_math_operations", "q6_number_memory", "q7_time_concept", "q8_counting_patterns",
    "q9_writing_quality", "q10_pencil_grip", "q11_word_spacing", "q12_copying_accuracy",
    "q13_social_cues", "q14_spatial_reasoning", "q15_routine_flexibility", "q16_nonverbal_comprehension",
    "q17_background_noise", "q18_verbal_instructions", "q19_sound_discrimination", "q20_listening_retention",
]
LABEL_COLS = ["dyslexia", "dyscalculia", "dysgraphia", "nvld", "apd"]

if not os.path.exists(MODEL_PATH) or not os.path.exists(DATA_PATH):
    print("Error: Model or Data file not found. Ensure paths are correct relative to the notebook location.")
else:
    model = joblib.load(MODEL_PATH)
    df = pd.read_csv(DATA_PATH)
    
    X = df[FEATURE_COLS]
    y_true = df[LABEL_COLS]
    y_pred = model.predict(X)
    
    print("✅ Model and Data loaded successfully.")
```

## 1. Classification Report
Detailed Precision, Recall, and F1-Scores for multi-label prediction.

```python
print(classification_report(y_true, y_pred, target_names=LABEL_COLS))
print(f"Global Hamming Loss: {hamming_loss(y_true, y_pred):.4f}")
```

## 2. Multi-Domain Confusion Matrices
Confusion matrices for each of the 5 learning difficulty domains.

```python
mcm = multilabel_confusion_matrix(y_true, y_pred)

for i, label in enumerate(LABEL_COLS):
    print(f"\n--- {label.upper()} ---")
    print("[[TN, FP],\n [FN, TP]]")
    print(mcm[i])
```
